```python
import json, time
from IPython.display import clear_output
from STservo import ST3215, Calibrator
with open('Controller.json', 'r') as fr: 
    config = json.load(fp=fr)
config
```

```python
id_arr=list(range(1, 13))
calib = Calibrator(port='/dev/ttyACM0', id_arr=id_arr, __set_posi_corr__=False)
for _id, _c in config.items(): 
    print(_id, _c)
    calib.__set_posi_corr__(dev_id=int(_id), step=_c['corr'], save=True)
```

```python
time.sleep(1)
try:
    while True:
        time.sleep(0.3)
        clear_output(wait=True)
        for _id in id_arr:
            status, corr = calib.__get_posi_corr__(dev_id=_id)
            print(f'id: {_id}, corr: {corr}, posi: {calib.readPosi(id_arr=[_id], debug=True)[_id]["posi"]}')  
except KeyboardInterrupt:
    pass
```
